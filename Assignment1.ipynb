```python
1. Concept of Standardization (Z-Score Normalization)
Standardization transforms numerical data to have a mean of 0 and a standard deviation of 1. This is achieved through z-score normalization, where each value is adjusted based on the distribution of the dataset.

The formula for standardization is:

𝑧
=
𝑥
−
𝜇
𝜎
z=
σ
x−μ
​


where:

𝑥
x = original value
𝜇
μ = mean of the variable
𝜎
σ = standard deviation of the variable
𝑧
z = standardized value
Why standardize?

Uniformity: Different features may have different ranges or units. Standardization puts all features on a similar scale.
Improved Performance: Many machine learning algorithms, especially those involving distance metrics (e.g., k-nearest neighbors, gradient descent), perform better when data is standardized.
Comparability: It helps in comparing features with different units or scales more effectively.
```

```python
#standarize numerical column
import pandas as pd

# Load your dataset
df = pd.read_csv('sales_data_with_discounts.csv')

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate mean and standard deviation
means = df[numerical_cols].mean()
std_devs = df[numerical_cols].std()

# Standardize the numerical columns
df_standardized = (df[numerical_cols] - means) / std_devs
```

```python
import seaborn as sns

# Plot box plots for numerical columns before and after standardization
for col in numerical_cols:
    plt.figure(figsize=(14, 6))

    # Original data
    plt.subplot(1, 2, 1)
    sns.boxplot(y=df[col])
    plt.title(f'Original Distribution of {col}')

    # Standardized data
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df_standardized[col])
    plt.title(f'Standardized Distribution of {col}')

    plt.tight_layout()
    plt.show()
```

```python
 Need for Converting Categorical Data into Dummy Variables (One-Hot Encoding)
Categorical Data and Machine Learning:

Machine learning models usually work with numerical data. Categorical data, which represents different categories or labels, needs to be converted into a numerical format for algorithms to process it.
One-hot encoding is a technique to transform categorical variables into a set of binary columns, where each column represents a distinct category. This ensures that the model can interpret the categorical data appropriately.
Why One-Hot Encoding?

Avoid Ordinal Implications: One-hot encoding avoids implying any ordinal relationship between categories (e.g., “Red”, “Green”, “Blue” are not inherently ordered).
Handle Non-Numeric Data: Many models require numerical input, and one-hot encoding provides a way to represent categorical data numerically without introducing arbitrary numbers.
```

```python
#apply one hot encoding to the categorical data

import pandas as pd

# Load dataset
df = pd.read_csv('sales_data_with_discounts.csv')
```

```python
#identify categorical columns

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
```

```python
# Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False)
```

```python
# Display the first few rows of the transformed dataset
print(df_encoded.head())

```

```python
import pandas as pd

# Example dataset
data = {
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red'],
    'Size': ['Small', 'Medium', 'Large', 'Medium', 'Large']
}

df = pd.DataFrame(data)

# Apply one-hot encoding
df_encoded = pd.get_dummies(df, drop_first=False)

# Display a portion of the transformed dataset
print(df_encoded.head())
```
